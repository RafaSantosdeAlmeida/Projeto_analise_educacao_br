```python
def calcular_correlacoes(df, target_var):
    """
    Calcula a correlação entre uma variável alvo e todas as outras variáveis numéricas em um DataFrame.
    
    Parâmetros:
    df (pd.DataFrame): DataFrame contendo os dados.
    target_var (str): Nome da variável alvo.
    
    Retorna:
    pd.Series: Correlações da variável alvo com as outras variáveis.
    """
    if target_var not in df.columns:
        raise ValueError(f"A variável '{target_var}' não está no DataFrame.")
    
    # Seleciona apenas colunas numéricas
    numeric_df = df.select_dtypes(include=['number'])
    
    if target_var not in numeric_df.columns:
        raise ValueError(f"A variável alvo '{target_var}' não é numérica.")
    
    # Calcula a correlação de Pearson
    correlacoes = numeric_df.corr()[target_var].drop(target_var)
    
    return correlacoes
```

```python
calcular_correlacoes(df_escolas[['TT_DIAS_FALTA_MEDICA',
       'TT_DIAS_FALTA_JUST', 'TT_DIAS_FALTA_INJUST', 'TT_DIAS_LIC_PREMIO',
       'TT_DIAS_LIC_GESTANTE', 'TT_DIAS_LIC_ACID_TRAB',
       'TT_DIAS_LIC_INTER_PARTIC', 'TOT_DIAS_AUSENCIAS', 'TOTAL_DIAS_MES', 'porc_ACERT_MAT']], 'porc_ACERT_MAT')
```
